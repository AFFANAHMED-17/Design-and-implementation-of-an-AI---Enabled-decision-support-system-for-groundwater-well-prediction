```python
import joblib

# load the lithology encoder used in Model-3
enc_lithology = joblib.load("./encoders/model_3_lithology_encoder.pkl")

# print mapping
lithology_mapping = {
    lith: int(code)
    for code, lith in enumerate(enc_lithology.classes_)
}

for lith, code in lithology_mapping.items():
    print(f"{code:02d}  ->  {lith}")
```
